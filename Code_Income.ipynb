```python
import pandas as pd
import csv
import gmaps
import numpy
```

```python
population_profile_df= pd.read_csv('neighbourhood-profiles-2016-csv.csv')
population_profile_df=population_profile_df.set_index('_id')
population_neigh_2016=population_profile_df.iloc[2,:]
population_neigh_2016.to_csv("Neighborhood Population.csv")
```

```python
incom_df=(population_profile_df.loc[1038:1053])
incom_df['Median Income']= [0,2500,5500,12500,22500,27500,32500,37500,42500,47500,55000,65000,75000,85000,95000,150000]


incom_df.iloc[:,4:(len(incom_df.T)-1)]=incom_df.iloc[:,4:(len(incom_df.T)-1)].replace('[^\d.]', '',regex=True).astype(float)
```

```python
neigh_inc=[]
for j in range(5,(len(incom_df.T)-1)):
    s=c=0
    for k in range(1,(len(incom_df)-1)):
        s=s+incom_df.iloc[k,j]*incom_df.iloc[k,(len(incom_df.T)-1)]
        c=c+incom_df.iloc[k,j]
    neigh_inc.append(s/c)
print(len(neigh_inc))
```

```python
neigh_income=['Income','Neighborhood Income','Calculated Data','Mean Income','']
for l in range(len(neigh_inc)):
    neigh_income.append(neigh_inc[l])  # appending two lists 
len(neigh_income)
neigh_income.append('')
income_df=incom_df.T
income_df['Neighborhood Mean Income']=neigh_income
income_df
```

```python
income_df.to_csv("Neighborhoods Mean Income.csv")
```
